### Create Lyric Data

```python
# Dependencies
import pandas as pd
import lyrics as l

df = pd.read_csv('Data/updated.csv', encoding='latin-1')
```

```python
df['album_title'].value_counts()
```

```python
# song_list with placeholders
song_list = [0] * df.shape[0]


# Loop through df to make list of dictionaries
for index, row in df.iterrows():
    
    # Create base dictionary
    dictionary = {
        'album': row['album_title'],
        'song': row['track_title'],
    }
    
    # Put wordcount of lyrics into dictionary
    try:
        lyrics = l.get_lyrics(row['track_title']).replace("\n", " ")
        words = len(lyrics.split(" "))
        dictionary['words'] = words
        print(index)
        
    # Put wordcount = 0 into dictionary 
    except:
        dictionary['words'] = 0
        print("fail: " + str(index))
    
    # Put dictionary into song_list
    song_list[index] = dictionary
```

```python
import json
with open('song_list2', 'w') as fout:
    json.dump(song_list, fout)
```
