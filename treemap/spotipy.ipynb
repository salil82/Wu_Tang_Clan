```python
import sys
import spotipy
import spotipy.util as util

scope = 'user-library-read'
album_id = 'spotify:album:3g8KDFecnLPa6YiBdHTl8K'
token = util.prompt_for_user_token(username,scope,client_id='89f83364a522438b96552f59ae43c3dd',client_secret='62643634002f4f34964b7c8fd48c0fef',redirect_uri='https://www.google.com/')

track_title = []
track_number = []
track_duration = []
album_uri_reference = []

for x in album_uri:

    if token:
        sp = spotipy.Spotify(auth=token)
        results = sp.album_tracks(x, limit=50, offset=0)
        album_tracks = results['items']
        while results['next']:
            results = spotify.next(results)
            album_tracks.extend(results['items'])

        for track in album_tracks:
            track_title.append(track['name'])
            track_number.append(track['track_number'])
            track_duration.append(track['duration_ms'])
            album_uri_reference.append(x)
            

    else:
        print("Can't get token for", username)
```
